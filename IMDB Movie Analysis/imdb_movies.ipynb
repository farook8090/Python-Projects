#### How do ratings vary by genre? Use a boxplot to visualize the differences in ratings across genres.

```python
df_genre_split = df[['score', 'genre']].copy()
df_genre_split = df_genre_split.dropna(subset=['score'])
df_genre_split['genre']=df_genre_split['genre'].str.split(",")
df_genre_split=df_genre_split.explode('genre')
df_genre_split['genre'] = df_genre_split['genre'].str.strip()
```

```python
#### Is there a correlation between the number of votes a budget and revenue? Create a scatter plot and calculate the correlation coefficient. What can you conclude?
```

```python
df_corr = df[['budget_x', 'revenue']].dropna()

correlation = df_corr['budget_x'].corr(df_corr['revenue'])

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_corr, x='budget_x', y='revenue', alpha=0.6)
plt.title(f'Scatter Plot of Budget vs Revenue\nCorrelation: {correlation:.2f}')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

# 6. Genre-Specific Analysis

#### Which genre has the highest average rating? Calculate the average rating for each genre and plot the results.

```python
df['genre'] = df['genre'].astype(str)
df_genre_rating = df[['score', 'genre']].copy()

df_genre_rating = df_genre_rating.dropna(subset=['score'])

df_genre_rating['genre'] = df_genre_rating['genre'].str.split(',')
df_genre_rating = df_genre_rating.explode('genre')
df_genre_rating['genre'] = df_genre_rating['genre'].str.strip()

avg_rating_by_genre = df_genre_rating.groupby('genre')['score'].mean().sort_values(ascending=False)

top_genre = avg_rating_by_genre.idxmax()
top_rating = avg_rating_by_genre.max()
# print(f"Genre with the highest average rating: {top_genre} ({top_rating:.2f})")

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rating_by_genre.values, y=avg_rating_by_genre.index, palette='mako')
plt.title('Average Movie Rating by Genre')
plt.xlabel('Average Rating (Score)')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

#### How does the popularity of genres vary over time? Plot the number of movies released per genre each year.

```python
df['year'] = df['date_x'].dt.year

df = df.dropna(subset=['year', 'genre'])

df['genre'] = df['genre'].astype(str)
df['genre'] = df['genre'].str.split(',')
df = df.explode('genre')
df['genre'] = df['genre'].str.strip()

genre_year_counts = df.groupby(['year', 'genre']).size().reset_index(name='movie_count')

genre_pivot = genre_year_counts.pivot(index='year', columns='genre', values='movie_count').fillna(0)

plt.figure(figsize=(14, 8))
sns.heatmap(genre_pivot.T, cmap='YlGnBu', linewidths=0.5)
plt.title('Number of Movies Released per Genre Each Year')
plt.xlabel('Year')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

# 7. Year and Trend Analysis

#### How has the average movie rating changed over the years? Plot the average rating for each year.

```python
df['year'] = df['date_x'].dt.year

df_ratings_by_year = df.dropna(subset=['year', 'score'])

avg_rating_per_year = df_ratings_by_year.groupby('year')['score'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_rating_per_year, x='year', y='score', marker='o')
plt.title('Average Movie Rating by Year')
plt.xlabel('Year')
plt.ylabel('Average Rating (Score)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Which years had the highest and lowest number of movie releases? Plot the number of movies released each year.

```python
df['year'] = df['date_x'].dt.year

df_years = df.dropna(subset=['year'])

movies_per_year = df_years['year'].value_counts().sort_index()

max_year = movies_per_year.idxmax()
min_year = movies_per_year.idxmin()
max_count = movies_per_year.max()
min_count = movies_per_year.min()

# print(f"Highest releases: {max_year} ({max_count} movies)")
# print(f"Lowest releases: {min_year} ({min_count} movies)")

plt.figure(figsize=(12, 6))
sns.barplot(x=movies_per_year.index, y=movies_per_year.values, color='skyblue')
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 8. Multivariate Analysis: Analyze multiple variables together.

#### Which genres are most popular in each decade? Create a bar plot showing the most frequent genres by decade.

```python
df['year'] = df['date_x'].dt.year

df = df.dropna(subset=['year', 'genre'])

df['decade'] = (df['year'] // 10 * 10).astype(int).astype(str) + 's'

df['genre'] = df['genre'].astype(str)
df['genre'] = df['genre'].str.split(',')
df = df.explode('genre')
df['genre'] = df['genre'].str.strip()

genre_decade_counts = df.groupby(['decade', 'genre']).size().reset_index(name='count')

top_genres_per_decade = genre_decade_counts.sort_values(['decade', 'count'], ascending=[True, False])\
                                           .groupby('decade').head(1)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_genres_per_decade, x='decade', y='count', hue='genre', dodge=False, palette='Set2')
plt.title('Most Popular Genre by Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Movies')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()
```

#### Plot a heatmap or pairplot to examine relationships between budget, revenue, scores.

```python
cols = ['budget_x', 'revenue', 'score']
df_corr = df[cols].dropna()

df_corr = df_corr.astype(float)

corr_matrix = df_corr.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Budget, Revenue, Score')
plt.tight_layout()
plt.show()
```

#### Are there specific genres or release years with higher-rated movies? Group by genre and year, then analyze the average rating.

```python
df['year'] = df['date_x'].dt.year

df = df.dropna(subset=['genre', 'year', 'score'])

df['genre'] = df['genre'].astype(str).str.split(',')
df = df.explode('genre')
df['genre'] = df['genre'].str.strip()

genre_year_rating = df.groupby(['genre', 'year'])['score'].mean().reset_index()

pivot_table = genre_year_rating.pivot(index='genre', columns='year', values='score')

plt.figure(figsize=(16, 10))
sns.heatmap(pivot_table, cmap='YlGnBu', linewidths=0.5, annot=False)
plt.title('Average Movie Ratings by Genre and Year')
plt.xlabel('Year')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```
