>***Customer Sentiment Analysis***

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

from selenium.webdriver.common.keys import Keys
import re
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.flipkart.com/apple-iphone-15-blue-128-gb/product-reviews/itmbf14ef54f645d?pid=MOBGTAGPAQNVFZZY&lid=LSTMOBGTAGPAQNVFZZYQRLPCQ&marketplace=FLIPKART")
```

```python
usernames_list = []
rating_list = []
review_text_list = []
```

```python
def data():
    usernames = driver.find_elements(By.XPATH,'//p[@class="_2NsDsF AwS1CA"]')
    ratings = driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    reviews = driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]/div/div')
    time.sleep(5)
    
    for username in usernames:
        name = username.text
        usernames_list.append(name)

    for rating in ratings:
        r = rating.text
        rating_list.append(r)

    for review in reviews:
        r = review.text
        review_text_list.append(r)
```

```python
# data scrapping
page_number = 1

for page_number in range(1,31):
    if page_number==1:
        data()
        nxt_button = driver.find_element(By.XPATH,'//a[@class="_9QVEpD"]')
        nxt_button.click()
        time.sleep(2)
    else:
        data()
        nxt_button = driver.find_elements(By.XPATH,'//a[@class="_9QVEpD"]')
        time.sleep(3)
        #print(nxt_button)
        for nxt_button in nxt_button:
            # print(nxt_button.text)
            if nxt_button.text == 'NEXT':
                nxt_button.click()
                time.sleep(2)
```

```python
print(f"Length of Usernames list is {len(usernames_list)}.")
print(f"Length of Ratings list is {len(rating_list)}.")
print(f"Length of Review Text List list is {len(review_text_list)}.")
print(f"Max length of all list is {max(len(usernames_list),len(review_text_list),len(review_text_list))}")
```

```python
# appending None if len of all three is not same.
differnce = max(len(usernames_list),len(review_text_list),len(review_text_list))-len(rating_list)
print(f"{differnce} number of None value to be inserted in rating_list.")
```

```python
for r in range(0,differnce):
    rating_list.append(None)
```

```python
print(f"Length of Usernames list is {len(usernames_list)}.")
print(f"Length of Ratings list is {len(rating_list)}.")
print(f"Length of Review Text List list is {len(review_text_list)}.")
```

```python
df = pd.DataFrame({
    "usernames":usernames_list,
    "reviews":review_text_list,
    "ratings":rating_list
})

df
```

> EDA

```python
df['usernames']=df['usernames'].astype('object')
df['reviews']=df['reviews'].astype('object')
df['ratings']=df['ratings'].astype('float32')
```

```python
print(f"DataFrame info before data cleaning.")
```

```python
df.info()
```

```python
df.isnull().sum()
```

> Data Cleaning

```python
# filling None values with median in ratings column.
df['ratings'] = df['ratings'].fillna(df['ratings'].median())
```

```python
print(f"DataFrame info after data cleaning.")
```

```python
df.isnull().sum()
```

```python
# review columns cleaning start
```

```python
df
```

```python
pattern = r"[a-zA-Z0-9]+"
review_text_cleaned_list = []

for review in df['reviews'].dropna().values:
    matches = re.findall(pattern, review)
    sentence = " ".join(matches)
    review_text_cleaned_list.append(sentence)
```

```python
df['reviews'] = review_text_cleaned_list
df
```

```python
# review columns cleaning end
```

>***Sentiment Analysis***

```python
#pip install textblob
```

```python
from textblob import TextBlob
```

```python
polarity_list = []
```

```python
subjectivity_list =[]
```

```python
len(review_text_cleaned_list)
```

```python
# Polarity analysis
```

```python
for review in review_text_cleaned_list:
    sentence = TextBlob(review)
    polarity=sentence.polarity
    polarity_list.append(round(polarity,2))
```

```python
# Subjectivity analysis
```

```python
for review in review_text_cleaned_list:
    sentence = TextBlob(review)
    subjectivity = sentence.subjectivity
    subjectivity_list.append(round(subjectivity,2))
```

```python
polarity_list
```

```python
subjectivity_list
```

```python
print(f"Length of Polarity is {len(polarity_list)}.")
print(f"Length of Subjectivity is {len(subjectivity_list)}.")
```

```python
df['Polarity'] = polarity_list
df['Subjectivity'] = subjectivity_list
```

```python
df
```

```python
df['Sentiment Type'] = df['Polarity'].apply(lambda x: "Positive" if x >= 0.1 else "Negative")
```

```python
df
```

```python
# Save as CSV
df.to_csv("Customer_Sentiment_Analysis.csv", index=False)

# Save as Excel
df.to_excel("Customer_Sentiment_Analysis.xlsx", index=False)
```

>***4. Data Analysis and Insights:***

```python
df = pd.read_csv("Customer_Sentiment_Analysis.csv")
df
```

```python
# df.groupby('Sentiment Type').size()

positive_count = (df["Sentiment Type"]=="Positive").sum()
negative_count = (df["Sentiment Type"]=="Negative").sum()
```

```python
print(f"Number of positive reviews {positive_count}.")
print(f"Number of negative reviews {negative_count}.")
```

```python
# Sentiment Distribution: Calculate the overall distribution of positive and negative sentiments for the 300 reviews.
```

```python
bar_colors = ['tab:blue','tab:orange']

plt.figure(figsize=(10, 4))
plt.bar(x=["Positive","Negative"],height=[positive_count,negative_count],label=["Positive","Negative"],color=bar_colors)
plt.title("Distribution of positive and negative sentiments for the 300 reviews")
plt.xlabel("Sentiment Type")
plt.ylabel("Counts")
plt.savefig("Distribution of positive and negative sentiments for the 300 reviews")
plt.show()
```

```python
# Average Rating vs Sentiment: Analyze if there is any correlation between the numeric ratings (1-5 stars) and sentiment polarity. Do higher ratings correspond with more positive sentiments?
```

```python
avg_polarity = df.groupby('ratings')['Polarity'].mean()
avg_polarity
```

```python
plt.figure(figsize=(12, 4))
plt.plot(avg_polarity.index, avg_polarity.values, marker='o')
plt.xlabel('Ratings (1‚Äì5 Stars)')
plt.ylabel('Average Sentiment Polarity')
plt.title('Average Sentiment Polarity vs. Rating')
plt.savefig("Average Sentiment Polarity vs Rating")
plt.show()
```

```python
# Word Cloud: Create a word cloud to identify the most frequently mentioned words in the positive and negative reviews.
```

```python
#pip install wordcloud
```

```python
from wordcloud import WordCloud
```

```python
positive_text = " ".join(df[df['Sentiment Type'] == "Positive"]['reviews'])
negative_text = " ".join(df[df['Sentiment Type'] == "Negative"]['reviews'])
```

```python
positive_wc = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
negative_wc = WordCloud(width=800, height=400, background_color='white').generate(negative_text)
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.imshow(positive_wc, interpolation='bilinear')
plt.axis('off')
plt.title("Positive Reviews Word Cloud")

plt.subplot(1, 2, 2)
plt.imshow(negative_wc, interpolation='bilinear')
plt.axis('off')
plt.title("Negative Reviews Word Cloud")

plt.show()
```

```python
# Review Length Analysis: Investigate if longer reviews are associated with more detailed sentiments, either positive or negative.
```

```python
df['Review_Length'] = df['reviews'].apply(lambda x: len(x.split()))
df.groupby('Sentiment Type')['Review_Length'].mean()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sentiment Type', y='Review_Length', data=df)
plt.title("Review Length vs Sentiment")
plt.show()
```
